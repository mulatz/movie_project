```python
import numpy as np
import pandas as pd
import difflib
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
data = pd.read_csv('movies.zip')
```

```python
data
```

###### summary of the data information 

```python
data.info()
```

###### description of the data by using describe() method

```python
data.describe()
```

###### add the new columens by using assign() method 

```python
data = data.assign(New_column = data['popularity']>21)
```

```python
data
```

###### now return random sample by using  sample() function method 

```python
data.sample(5)
```

###### remove rows or columns by using drop()method function 

```python
data = data.drop(columns=['popularity','original_title'])
```

```python
data
```

###### remove any values (rows) that contain missing values dropna() method function 

```python
data.dropna(subset = 'homepage')
```

###### number of rows and columns in the data frame

```python
data.shape
```

###### selecting the relevant features for recommendation

```python


selected_features = ['genres','keywords','tagline','cast','director']
print(selected_features)
```

```python
# replacing the null valuess with null string
```

```python


for feature in selected_features:
 data[feature] = data[feature].fillna('')
```

###### combining all the 5 selected features

```python


combined_features = data['genres']+' '+data['keywords']+' '+data['tagline']+' '+data['cast']+' '+data['director']
```

```python
print(combined_features)
```

###### converting the text data to feature vectors

```python


vectorizer = TfidfVectorizer()
```

```python
feature_vectors = vectorizer.fit_transform(combined_features)
```

```python
print(feature_vectors)
```

###### now  getting the similarity scores using cosine similarity

```python

similarity = cosine_similarity(feature_vectors)
```

```python
print(similarity)
```

###### let's see the shape of the data by using shape()method function

```python
print(similarity.shape)
```

###### getting the movie name from the user

```python

movie_name = input(' Enter your favourite movie name : ')
```

###### now time to  creating a list with all thes given in the dataset movie name

```python

list_of_all_titles = data['title'].tolist()
print(list_of_all_titles)
```

###### finding the close match for the movie name given by the user

```python


find_close_match = difflib.get_close_matches(movie_name, list_of_all_titles)
print(find_close_match)
```

```python
close_match = find_close_match[0]
print(close_match)
```

###### now finding the index of the movie with title

```python

index_of_the_movie = data[data.title == close_match]['index'].values[0]
print(index_of_the_movie)
```

###### getting a list of similar movies

```python

similarity_score = list(enumerate(similarity[index_of_the_movie]))
print(similarity_score)
```

```python
len(similarity_score)
```

##### sorting the movies based on their similarity score

```python

sorted_similar_movies = sorted(similarity_score, key = lambda x:x[1], reverse = True) 
print(sorted_similar_movies)
```

###### print the name of similar movies based on the index

```python

print('Movies suggested for you : \n')

i = 1

for movie in sorted_similar_movies:
  index = movie[0]
  title_from_index = data[data.index==index]['title'].values[0]
  if (i<30):
    print(i, '.',title_from_index)
    i+=1
```
